```python
# import libraries
import pandas as pd
from sklearn.neighbors import NearestNeighbors
```

```python
# read the data
movies_df = pd.read_csv('https://github.com/ArinB/CA05-kNN/raw/master/movies_recommendation_data.csv')
```

```python
# show the first 5 rows of the dataframe
movies_df.head()
```

```python
# show the number of rows and columns
movies_df.shape
```

```python
# drop the label column
movies_df = movies_df.drop('Label', axis =1)
movies_df.head()
```

```python
# get movie names for future reference
movie_names = movies_df[['Movie ID', 'Movie Name']]
movie_names.head()
```

```python
# get the data for predictions
movies_data = movies_df.iloc[:, 2:10]
movies_data.head()
```

# Building a recommender system

```python
# create the values for the movie The Post
the_post = [7.2, 1, 1, 0, 0, 0, 0, 1]
```

```python
# create and fit the model, specify the number of neighbors
neigh = NearestNeighbors(n_neighbors = 5)
neigh.fit(movies_data)
```

```python
# print 5 similar movies to 'The Post'
print('You might like these movies:')
print()
for index in neigh.kneighbors([the_post])[1][0]:
  print(movie_names.iloc[index,1])
print()
print("Don't forget your popcorn!")
```
